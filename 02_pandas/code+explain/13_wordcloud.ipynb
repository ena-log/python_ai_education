```python
import random
```

```python
# 워드 클라우드를 그레이톤으로 표시하는 함수
# word => 워드 클라우드로 표시할 단어
# font_size => 워드 클라우드로 표시할 단어의 크기
# position => 워드 클라우드로 표시할 단어의 위치
# **hwargs => 워드 클라우드로 표시할 단어의 글꼴이 저장된 경로
def gray_color_func(word, font_size, position, orientation, random_state=None, **hwargs):
    # print(word, font_size, position, orientation, hwargs)
    return 'hsl(0, 0%%, %d%%)' % random.randint(60, 100)
```

```python
plt.figure(figsize=(15, 10))
plt.imshow(wordcloud.recolor(color_func=gray_color_func, random_state=3), interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
import nltk
from konlpy.corpus import kolaw
from konlpy.tag import Twitter
t = Twitter()
```

```python
# konlpy 패키지는 대한민국 헌법 말뭉치 kolaw와 국회법안 말뭉치인 kobill을 제공한다.
# constitution.txt는 파일 데이터 없이 파일 이름만 지정하면 읽어올 수 있다.
ko_con_text = kolaw.open('constitution.txt').read()
ko_con_text
```

```python
# 한글 데이터 형태소 분석을 실행한다.
tokens_ko = t.nouns(ko_con_text)
tokens_ko
```

```python
# 불용어를 정의한다.
stop_words = ['제', '월', '일', '조', '수', '때', '그', '이', '바', '및', '안', '정']
tokens_ko = [each_word for each_word in tokens_ko if each_word not in stop_words]
```

```python
ko = nltk.Text(tokens_ko)
ko.vocab().most_common(50)
```

```python
data = ko.vocab().most_common(500)
tmp_data = dict(data)
tmp_data
```

```python
# 딕셔너리에 저장된 데이터는 generate_from_frequencies() 메소드를 사용해서 워드 클라우드를 실행해야 한다.
wordcloud = WordCloud(
    font_path='c:/windows/fonts/나눔고딕코딩.ttf',
    background_color='white',
    relative_scaling=0.5
).generate_from_frequencies(tmp_data)
plt.figure(figsize=(15, 10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
