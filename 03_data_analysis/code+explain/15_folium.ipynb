```python
# folium은 Leaflet.js를 이용해서 위치 좌표계 값을 지도위에 시각화 할 수 있다.
# !pip install folium
import folium
# 보고자 하는 위치의 좌표계를 얻어내는 가장 단순한 방법은 구글맵을 이용하는 방법이다.
```

```python
# folium 라이브러리의 Map() 메소드의 location 속성으로 지정한 위치를 중심으로 하는 zoom_start 옵션으로 지정한 배율을 사용하는
# 지도를 만든다.
# folium.Map(location=[위도, 경도], zoom_start=배율[, tiles='지도 스타일'])
map_folium = folium.Map(location=[37.5662952, 126.9757511], zoom_start=15)

# folium의 save() 메소드로 작성된 지도 차트를 html 파일로 저장할 수 있다.
map_folium.save('./output/map.html')
map_folium
```

```python
map_folium = folium.Map(location=[37.5662952, 126.9757511], zoom_start=16)

# folium 라이브러리의 Marker() 메소드의 location 속성으로 지정한 위치에 마커를 만든다.
# popup 속성으로 Marker를 클릭했을 때 표시할 문자열을 지정할 수 있다.
# 작성된 Marker는 add_to() 메소드를 사용해서 지도에 추가한다.
folium.Marker(location=[37.56672038822997, 126.977945098114], popup='서울특별시청').add_to(map_folium)
folium.Marker(location=[37.56600047722329, 126.97515682704652], popup='덕수궁').add_to(map_folium)
folium.Marker(location=[37.57500644389687, 126.9719235827721], popup='서울지방경찰청').add_to(map_folium)
folium.Marker(location=[37.57622235345115, 126.97691299811414], popup='광화문').add_to(map_folium)
folium.Marker(location=[37.57002498561249, 126.98231609811403], popup='영풍문고 - 종로본점').add_to(map_folium)

map_folium.save('./output/map.html')
map_folium
```

```python
map_folium = folium.Map(location=[37.5662952, 126.9757511], zoom_start=16)

# Marker를 클릭했을 때 팝업되는 말풍선의 크기가 마음에 들지 않으면 Popup() 메소드의 max_width 속성으로 변경하면 된다.
popup = folium.Popup('서어어어어우우우울특벼어어어어얼시처어어어어어엉', max_width=500)

folium.Marker(location=[37.56672038822997, 126.977945098114], popup=popup).add_to(map_folium)
folium.Marker(location=[37.56600047722329, 126.97515682704652], popup='덕수궁').add_to(map_folium)
folium.Marker(location=[37.57500644389687, 126.9719235827721], popup='서울지방경찰청').add_to(map_folium)
folium.Marker(location=[37.57622235345115, 126.97691299811414], popup='광화문').add_to(map_folium)
folium.Marker(location=[37.57002498561249, 126.98231609811403], popup='영풍문고 - 종로본점').add_to(map_folium)

map_folium.save('./output/map.html')
map_folium
```

```python
map_folium = folium.Map(location=[37.5662952, 126.9757511], zoom_start=16)

# icon 속성에 Icon() 메소드로 마커의 색상과 아이콘 모양을 지정할 수 있다.
# Marker위에 표시하는 아이콘은 https://getbootstrap.com/docs/3.3/components/#glyphicons-glyph에서 제공하는 이미지를 표시할
# 수 있다.
folium.Marker(location=[37.56672038822997, 126.977945098114], 
              icon=folium.Icon(color='red', icon='save')).add_to(map_folium)
folium.Marker(location=[37.56600047722329, 126.97515682704652], 
              icon=folium.Icon(color='green', icon='send')).add_to(map_folium)
folium.Marker(location=[37.57500644389687, 126.9719235827721], 
              icon=folium.Icon(color='purple', icon='star')).add_to(map_folium)
folium.Marker(location=[37.57622235345115, 126.97691299811414], 
              icon=folium.Icon(color='blue', icon='music')).add_to(map_folium)
folium.Marker(location=[37.57002498561249, 126.98231609811403], 
              icon=folium.Icon(color='orange', icon='zoom-out')).add_to(map_folium)

map_folium.save('./output/map.html')
map_folium
```
