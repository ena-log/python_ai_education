```python
import warnings
warnings.filterwarnings("ignore")
import requests
```

시도 코드 크롤링

```python
targetSite = 'https://www.starbucks.co.kr/store/getSidoList.do' # 타겟 사이트, 요청 경로
# Request URL(https://www.starbucks.co.kr/store/getSidoList.do)을 요청하는 방식인 Request Method가 POST 방식이므로 requests
# 모듈의 post() 메소드로 서버에 정보를 요청한다.
# Request Method가 POST 방식일 경우 서버에 정보를 요청할 때 넘겨주는 데이터가 있으므로 Form Data의 내용을 같이 전달해야 한다.
request = requests.post(targetSite, data={
    'rndCod': '9I6I6ODNDQ'
})
print(request)
print(type(request.text))
print(request.text)
```

```python
# json 모듈의 loads() 메소드로 크롤링한 json 타입의 문자열을 파이썬에서 처리하기 위해 딕셔너리 타입으로 변환한다.
import json
star_json = json.loads(request.text) # json 타입의 문자열을 딕셔너리로 변환한다.
print(type(star_json))
print(star_json)
```

```python
# requests 모듈의 json() 메소드로 크롤링한 json 타입의 문자열을 파이썬에서 처리하기 위해 딕셔너리 타입으로 변환한다.
sidoList = request.json() # json 타입의 문자열을 딕셔너리로 변환한다.
print(type(sidoList))     # <class 'dict'>
print(len(sidoList))      # 1
print(sidoList)
```

```python
print(type(sidoList['list'])) # <class 'list'>
print(len(sidoList['list']))  # 17
print(sidoList['list'])
```

```python
print(type(sidoList['list'][0])) # <class 'dict'>
print(len(sidoList['list'][0]))
print(sidoList['list'][0])
```

```python
# 시도 코드의 개수 만큼 반복하며 시도 코드('sido_cd')를 key로 하고 시도 이름('sido_nm')을 value로 하는 딕셔너리를 만든다.
sido_List = dict()
for sido in sidoList['list']:
    # print('{} : {}'.format(sido['sido_cd'], sido['sido_nm']))
    # sido_List 딕셔너리에 시도 코드를 key로 하고 시도 이름을 value로 해서 저장시킨다.
    sido_List[sido['sido_cd']] = sido['sido_nm']
print(sido_List)
```

구군 코드 크롤링

```python
# 크롤링할 시도 코드를 입력받는다.
sido_cd = input('구군 코드를 검색할 시도 코드를 입력하세요 : ')
targetSite = 'https://www.starbucks.co.kr/store/getGugunList.do'
request = requests.post(targetSite, data={
    'sido_cd': sido_cd
})
gugunList = request.json()
# print(gugunList['list'][0])

# 구군 코드의 개수 만큼 반복하며 코드('gugun_cd')를 key로 하고 시도 이름('gugun_nm')을 value로 하는 딕셔너리를 만든다.
gugun_List = dict()
for gugun in gugunList['list']:
    # print('{} : {}'.format(gugun['gugun_cd'], gugun['gugun_nm']))
    gugun_List[gugun['gugun_cd']] = gugun['gugun_nm']
print(gugun_List)
```

```python
# 크롤링할 구군 코드를 입력받는다.
# gugun_cd = input('매장 정보를 확인할 구군 코드를 입력하세요 : ')

targetSite = 'https://www.starbucks.co.kr/store/getStore.do?r=R66KCOMOB6'
request = requests.post(targetSite, data={
    'ins_lat'   : 37.5108295,         # 위도
    'ins_lng'   : 127.02928809999999, # 경도
    'p_sido_cd' : '',                 # 시도 코드
    'p_gugun_cd': '0118',             # 구군 코드
    'in_biz_cd' : '',                 # ?
    'iend'      : 1600,               # 서버가 응답하는 최대 매장의 개수
    'set_date'  : ''                  # ?
    # 'in_biz_cd', 'set_date'를 넘겨주지 않으면 서버에서 데이터를 얻어오지 못한다.
})
storeList = request.json()
print(len(storeList['list']))

count = 0
dt = 0
for store in storeList['list']:
    count += 1
    print('{0:4d}. {1} : {2}({3}, {4})'.format(count, store['s_name'], store['doro_address'], store['lat'], store['lot']))
    if store['s_name'].find('DT') >= 0:
        dt += 1
print('DT 개수 : {}'.format(dt))
```
