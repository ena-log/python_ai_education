```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def arr_print(arr):
    print('type: {}'.format(type(arr)))
    print('shape: {}, dimension : {}, dtype : {}'.format(arr.shape, arr.ndim, arr.dtype))
    print('numpy 배열의 데이터\n', arr)
```

### 1. numpy의 linspace() 메소드 사용

```python
a = np.linspace(0, 1, 5)
arr_print(a)
```

```python
plt.plot(a)
plt.show()
```

### 2. numpy의 arange() 메소드 사용

```python
a = np.arange(0, 10, 2, dtype=np.float32)
arr_print(a)
```

```python
plt.plot(a)
plt.show()
```

### 3. numpy의 logspace() 메소드 사용

```python
a = np.logspace(0.1, 1, 20)
arr_print(a)
```

```python
plt.plot(a , 'o')
plt.show()
```
